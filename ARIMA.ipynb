# ARIMA

```python
import numpy as np #importing all the moddels needed
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa import seasonal,arima_model
import statsmodels.api as st
```

```python
data = pd.read_csv("./monthly-milk-production-pounds-p.csv") #importing the dataset
```

```python
data.head() 
```

### Data preprocessing

```python
data.dropna(axis = 0,inplace = True) #deleting all the null values
```

```python
data.columns = ["mon","mp"] #Setting the column names
```

```python
data["mon"]=pd.to_datetime(data["mon"]) #setting complete date 
```

```python
data.set_index("mon",inplace = True) #making date as index values
```

```python
data.head()
```

```python
tm = data["mp"] 
```

```python
tm1 = tm.copy()
```

```python
vd = seasonal.seasonal_decompose(tm1) #to recognize the pattern present in the data
```

```python
vd.plot()
```

### Predicting values

```python
num_sample_to_forcast = 24 #24 represents 2 years
```

```python
for i in range(0,num_sample_to_forcast):
    model1 = st.tsa.statespace.SARIMAX(tm,order=(0,1,1),seasonal_order = (1,1,1,12)) #defining the model
    res1 = model1.fit() #traning the model
    new_op = res1.forecast() #predicting the values
    print(new_op)
    tm = pd.concat([tm,new_op])
```

```python
tm.plot() #plotting old data
tm1.plot() #plotting predicted data
plt.show()
```
